#ACTIVIDAD 2.2 (Extracción de Datos)

```python
#Convertir en dataframe
df= pd.read_csv('NewOrleans.csv')
df
```

a) Los registros de 5 “host name” diferentes

```python
#a) Los registros de 5 “host name” diferentes
cinco_hosts = df["property_type"].unique()
cinco_hosts
#a_result = df[df["property_type"].isin(cinco_hosts)]
#a_result.head()
```

```python
# Elige 5 tipos de propiedad específicos manualmente
tipos_elegidos = ["Entire rental unit", "Entire guest suite", "Private room in home", "Room in boutique hotel", "Entire townhouse"] 

# Filtra el DataFrame para mostrar solo las observaciones que coincidan con esos tipos
a_result = df[df["property_type"].isin(tipos_elegidos)]

# Muestra todas las observaciones filtradas (no solo .head())
a_result
```

```python
#Analizar categorias de una columna
df1 = a_result.groupby(['property_type'])['property_type'].count()
df1
```

```python
f1=a_result.to_csv("f1.csv", index=False)
```

b) Los registros de los host que se hallan unido a Airbnb “despues del año 2020”

```python
# b) Host_since > 2020
df["host_since"] = pd.to_datetime(df["host_since"], errors="coerce")
b_result = df[df["host_since"] >= '2021-01-01']
b_result 
```

```python
df.info()
```

```python
df2 = b_result.groupby(['host_since'])['host_since'].count()
df2
```

```python
f2=b_result.to_csv("f2.csv", index=False)   
```

c) Los registros de los host que responden “a más tardar en 1 día”

```python
df3 = df.groupby(['host_response_time'])['host_response_time'].count()
df3
```

```python
# c) Respuesta rápida (en 1 día o menos)
rapidas = ["within an hour", "within a few hours", "within a day"]
c_result = df[df["host_response_time"].str.lower().isin(rapidas)]
c_result
```

```python
f3=c_result.to_csv("f3.csv", index=False)
```

d) Los registros de las filas “Múltiplos de 200… (400, 600, 800, etc.)”

```python
#Filas incrementales en un fator
filas= df.iloc[0::200,:] #Filas con incrementos de 100 a partir de la fila 0 
filas
```

```python
f4=filas.to_csv("f4.csv", index=False)
```

e) Los registros de los tipos de cuarto: “Entire home” y “Private room”

```python
df4 = df.groupby(['room_type'])['room_type'].count()
df4
```

```python
# e) Filtrar tipos de cuarto específicos
permitidos = ["entire home/apt", "private room"]
e_result = df[df["room_type"].str.lower().isin(permitidos)]
e_result
```

```python
f5=e_result.to_csv("f5.csv", index=False)
```

f) Los registros que “a lo mucho cuenten con 4 camas”

```python
df5 = df.groupby(['beds'])['beds'].count()
df5
```

```python
# f) Filtro por número de camas
df["beds"] = pd.to_numeric(df["beds"], errors="coerce")
f_result = df[df["beds"] <= 4]
f_result
```

```python
f6=f_result.to_csv("f6.csv", index=False)
```
